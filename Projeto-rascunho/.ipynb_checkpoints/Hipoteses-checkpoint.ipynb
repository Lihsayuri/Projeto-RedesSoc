# Introdução


Antes de começar as hipóteses, primeiro será relembrado o contexto e as hipóteses com as quais estamos trabalhando. Dessa forma, em relação ao contexto, como definido no início do trabalho, tem-se que engloba o fluxo de migração dos refugiados ou pessoas buscando asilo do ano de 2000 a 2016 excluindo as migrações que ocorrem de uma região a outra de um mesmo país.

Para esse contexto, estabelecemos alguns conceitos muito importantes:

- Estado crítico: indica se o país está em guerra ou em algum conflito.
- Abertura : indica se o país de destino aceita mais facilmente os refugiados do que outros, isto é, se todo o processo de admissão é menos burocrático e consequentemente mais pessoas são aceitas.
- Intermediação : se um país recebe muitos refugiados pois está em uma posição na rede que o favorece a isso, por exemplo: entre dois países em guerra.
- Centro : se um país possui um fluxo tanto de entrada quanto de saída considerável em relação aos demais países da rede ele pertencerá ao centro.
- Desenvolvimento : é o desenvolvimento socioeconômico de um país.
- Localização : são países com proximidade geográfica a outros países - os países vizinhos.
- Vazão de pessoas: é o fluxo de saída de pessoas de um país.
- Chegada de pessoas: é o fluxo de entrada de pessoas em um país.
- Contenção populacional : determina o quanto um país restringe seus cidadãos a permanecerem ali ou dificulta muito o processo de deixar o próprio país.


Através desses conceitos, 6 hipóteses foram formuladas e o objetivo deste documento é testar essas hipóteses para descobrir se há relações entre os conceitos e se serão ou não corroboradas de acordo com o que foi proposto.

Primeiro vamos começar carregando a base de dados que foi construída levando em consideração todos os conceitos acima.

## HIPÓTESE A [CRIT, CORE, CONP = EXT ]

A primeira hipótese a ser testada envolve dois conceitos: estado crítico e vazão de pessoas. Foi proposto que: quanto mais crítico é o estado de um país, maior será a vazão de pessoas desse país. E com essa hipótese, o seguinte mecanismo foi construído:

"Quando um país está passando por alguma guerra, conflitos ou por um momento economicamente e socialmente difícil, as pessoas buscam meios para encarar esses problemas. Alguns permanecem no país onde moram e lidam com as consequências disso, mas outros tomam a violência, as guerras e conflitos como motivação para sair de suas cidades e países de origem, ou seja, veêm a imigração como uma opção a ser considerada. Dessa forma, as pessoas buscam meios de sair de seu país local para um outro país e o fazem para escapar da situação crítica em que estão vivendo -  seja isso de maneira legal ou ilegal - , o que acarreta em um maior fluxo de saída"



E para testar essa hipótese, as seguintes variáveis foram selecionadas: 
- n_wars: número de guerras de um país (estado crítico - CRIT);
- coreness_discreto (centro - CORE);
- conp (contenção populacional - CONP);
- out_degree (vazão de pessoas - EXT).  

Vale ressaltar que as variáveis: coreness_discreto e conp são variáveis de controle nesse caso, isso porque :
- Quanto mais pessoa saem de um país, mais ao centro ele estará - já que o centro contempla países com grande fluxo de entrada e saída de pessoas.
- Quanto mais restringidos são os vistos de saída de um país, menos pessoas sairão dele (exemplo: Coreia do Norte). 

Essas variáveis não são o foco nessa análise, mas por terem relação com a variável dependente, precisam ser consideradas na regressão para não influenciar o resultado final. 

A regressão nesse caso será linear, pois embora o out_degree seja uma variável discreta, ela possui uma imensa gama de valores, não ficando restrita a poucos valores como 0 ou 1.

```python
result_hip1 = reg.linear(data=data, formula='out_degree ~ n_wars + coreness + pop_cont')
result_hip1.summary()
```

```python
result_hip1.nano_summary()
```

Como pode ser visto, a hipótese nesse teste corroborou. Isso porque o p value está significativo (p < 0.01) e o coeficiente é positivo, indicando uma relação proporcional. 

Mas antes vamos analisar o gráfico de resíduos para garantir que os erros são aleatórios.

### Gráfico de resíduos

```python
result_hip1.plot_residuals()
```

Como as variáveis independentes: número de guerras, coreness e contenção populacional são discretas, o gráfico de resíduos apresenta uma disposição de resultados um pouco espaçada e concentrada em alguns pontos - como se formassem faixas. E isso pode ser demonstrada mais detalhadamente quando plotamos o gráfico de resíduos para cada uma das variáveis independentes de maneira individual:  

#### Gráfico de resíduos do CRIT 

```python
result_hip1.plot_residuals(data['n_wars'])
```

#### Gráfico de resíduos do CORE

```python
result_hip1.plot_residuals(data['coreness'])
```

#### Gráfico de resíduos do CONT

```python
result_hip1.plot_residuals(data['pop_cont'])
```

Mas isso não é um problema e nem algo que parece denotar um padrão aos erros. Sendo assim, a hipótese foi corroborada de fato. Logo: quanto mais crítico é o estado de um país, mais pessoas saem dele.

## Hipótese B [GEO = CORE ]

A segunda hipótese a ser testada envolve dois conceitos: a localização e o coreness. Foi proposto que: quanto mais próximo um país é de outros e melhor posicionado geograficamente ele está (isso é, possui muitos países ao seu redor não ficando isolado), mais ao centro esse país estará. E com essa hipótese o seguinte mecanismo foi construído:

"Quanto mais localizado ao centro um país está, ou seja, quanto maior o fluxo entre os países ao seu redor, como é o caso de países no centro europeu, mais maneiras de se chegar a ele irão existir. Com mais maneiras de se chegar, mais fácil será de ocorrer vazão e chegada de pessoas, justamente por estarem posicionados melhor geograficamente e mais próximos de diversos outros países e/ou continentes. Além disso, países como ilhas, por estarem mais distantes entre si e mais isoladas geograficamente, tornam a chegada até lá mais difícil - sendo os únicos meios de chegar até a ilha avião ou navio. Dessa forma, nesse casos, as ilhas tendem a estar fora do centro. Logo, quanto melhor localizado o país, mais ao centro ele estará."


E para testar essa hipótese, as seguintes variáveis foram selecionadas: 
- coreness_discreto (centro - CORE);
- n_neighbours (localização - GEO);
- out_degree (vazão de pessoas - EXT).  

A regressão nesse caso será logística, e isso porque a variável dependente : o coreness é uma variável discreta adotando como possíveis valores apenas 0 ou 1.

```python
result_hip2 = reg.logistic(data=data, formula='coreness_discreto ~ n_neightbours')
result_hip2.summary()
```

```python
result_hip2.nano_summary()
```

A princípio, mais uma vez, a hipótese parece ter sido corroborada, já que o p value é significativo (p < 0.01) e o coeficiente é positivo, indicando uma relação proporcional.

Mas antes, mais uma vez é preciso analisar o gráfico de resíduos.

### Gráfico de resíduos

```python
result_hip2.plot_residuals()
```

Como pode ser visto, os erros possuem aleatoriedade e não apresentam nenhum padrão evidente. Assim, não é necessário manipular a regressão e a hipótese foi corroborada.

Logo, quanto mais países vizinhos um país tem, mais ao centro ele estará.

## Hipótese H [OPEN, EXT, CORE = IN]

A oitava hipótese a ser testada envolve quatro conceitos: a abertura, a vazão de pessoas, o coreness e a chegada de pessoas. Foi proposto que: quanto maior a abertura de um país, mais pessoas entram nele. E com essa hipótese o seguinte mecanismo foi construído:

"Países que possuem um processo de aceitação de refugiados menos burocrático, fazem com que pessoas que estão procurando novos lares fiquem de olho nesses países. Dependendo do quão crítica a situação é, alguns simplesmente vão de qualquer jeito para o país e depois tentam dar entrada em algum processo legal de acolhimento. Assim, se a pessoa busca sair do país o quanto antes, é mais provável que ela busque aqueles países que possuem maior abertura, políticas de incentivo à imigração e que já recebem muitas pessoas, a fim de, após sua entrada na nação, também consiga se manter e fazer parte do mercado de trabalho daquele lugar."

E para testar essa hipótese, as seguintes variáveis foram selecionadas:

- open (abertura - OPEN);
- coreness_discreto  (coreness - CORE);
- out_degree (vazão de pessoas - EXT);
- in_degree (chegada de pessoas -IN).

Vale ressaltar que as variáveis: coreness_discreto e out_degree são variáveis de controle nesse caso, isso porque :

- Quanto mais pessoa saem ou entram em um país, mais ao centro ele estará - já que o centro contempla países com grande fluxo de entrada e saída de pessoas.
- Quanto mais pessoas saem de um país, mais pessoas vão chegar em outros países.

A regressão nesse caso será linear, pois embora o in_degree seja uma variável discreta, ela possui uma imensa gama de valores, não ficando restrita a poucos valores como 0 ou 1.



```python
result_hip8= reg.linear(data=data, formula='in_degree ~ open + out_degree + coreness')
result_hip8.summary()
```

```python
result_hip8.nano_summary()
```

A princípio a hipótese foi corroborada, já que o p value é baixo (p < 0.05) e o coeficiente é positivo - o que indica uma relação direta entre os dois conceitos de centro e chegada de pessoas.

Agora vamos analisar o gráfico de resíduos.

### Gráfico de resíduos

```python
result_hip8.plot_residuals()
```

Como visto, os pontos do gráfico parecem estar bem "esmagados" em y, o que torna a visualização difícil - sem ser possível afirmar com certeza se os erros apresentam aleatoriedade ou não. Dessa forma, podemos aplicar um logaritmo à variável independente, como foi feito abaixo:

```python
result_hip8= reg.linear(data=data, formula='np.log1p(in_degree) ~ open + out_degree + coreness')
result_hip8.summary()
```

```python
result_hip8.nano_summary()
```

```python
result_hip8.plot_residuals()
```

Agora sim podemos ver melhor os pontos do gráfico e notar que de fato os erros apresentam aleatoriedade.

Logo, a hipótese foi corroborada e podemos afirmar que: quanto maior a abertura de um país, mais pessoas chegam até ele.
