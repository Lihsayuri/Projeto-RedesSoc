```python
import numpy as np
import pandas as pd
import seaborn as sns
import regression as reg

from scipy import stats
from matplotlib import pyplot as plt

import openpyxl
```

## CRIT, CORE, CONP = EXT 

Crit: Estado crítico -> número de guerras.

Core: Coreness.

Conp: Contenção populacional

Ext: peso do out degree.

```python
data = pd.read_excel('./final_db.xlsx')
data
```

```python
data.info()
```

```python
result_hip1 = reg.linear(data=data, formula='out_degree ~ n_wars + coreness + pop_cont')
result_hip1.summary()
```

```python
result_hip1.nano_summary()
```

```python
#sns.regplot(x=data['coreness'], y= np.log(data['out_degree']));
```

### Gráfico de resíduos

```python
result_hip1.plot_residuals()
```

### OPEN, EXT, CORE = IN

OPEN: Abertura de um país (número de rejeições)

EXT : peso do out degree.
    
Core : coreness

IN: peso do in-degree

```python
result_hip8= reg.linear(data=data, formula='in_degree ~ rejected + out_degree + coreness')
result_hip8.summary()
```

```python
result_hip8.nano_summary()
```

### Gráfico de resíduos

```python
result_hip8.plot_residuals()
```
