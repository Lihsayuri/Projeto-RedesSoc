```python
import pandas as pd
```

```python
df = pd.read_csv('./raw-files/asylum_seekers.csv')
```

```python
df
```

```python
df_filtrado = df.loc[:, ["Year", "Country / territory of asylum/residence", "decisions_recognized", "Origin"]]
```

```python
df_filtrado = df_filtrado.rename(columns={"Year":"year", 
                            "Country / territory of asylum/residence": "destiny",
                            "decisions_recognized": "accepted",
                            "Origin": "origin"})
```

```python
df_filtrado = df_filtrado.fillna(0)
```

```python
df_filtrado
```

```python
df_filtrado["accepted"] = df_filtrado.accepted.replace('*','0')
```

```python
df_filtrado
```

```python
df_filtrado["accepted"] = pd.to_numeric(df_filtrado.accepted)
```

```python
df_agg = df_filtrado.groupby(by=["year", "destiny", "origin"]).sum()
```

```python
df_agg = df_agg.reset_index()
```

```python
df_agg
```

```python
df_agg_2001 = df_agg.loc[df_agg.year == 2001, :]
df_agg_2001.to_csv("df_agg_2001_accepted.csv")
```

```python
df_agg_2008 = df_agg.loc[df_agg.year == 2008, :]
df_agg_2008.to_csv("df_agg_2008_accepted.csv")
```

```python
df_agg_2009_2010 = df_agg.loc[(df_agg.year == 2009) | (df_agg.year == 2010), :]
df_agg_2009_2010
```

```python
df_agg_2009_2010.groupby(["destiny", "origin"]).sum().reset_index().loc[:,["destiny", "accepted", "origin"]]
```

```python
df_agg_2009_2010.to_csv("df_agg_2009_2010_accepted.csv")
```
