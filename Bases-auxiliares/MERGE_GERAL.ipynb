```python
import pandas as pd
```

```python
df_rejected = pd.read_csv('./output-files/df_rejected.csv', index_col=0)
df_rejected.rename(columns={"destiny":"Country"}, inplace=True)
df_rejected
```

```python
df_general = pd.read_excel('./output-files/IDH_war_neighbor.xlsx')
df_general
```

```python
df2 = df_general.merge(df_rejected, on="Country", how="left")
```

```python
df2.rename(columns={"Country":"country"}, inplace=True)
```

```python
df2
```

```python
df_betweenness = pd.read_excel('./metrics/betweenness.xlsx', index_col=0)
df_CONP = pd.read_excel('./metrics/CONP.xlsx', index_col=0)
df_coreness = pd.read_excel('./metrics/coreness.xlsx', index_col=0)
df_fluxo_entrada = pd.read_excel('./metrics/fluxo_entrada.xlsx', index_col=0)
df_fluxo_saida = pd.read_excel('./metrics/fluxo_saida.xlsx', index_col=0)
```

```python
df_betweenness = df_betweenness.reset_index()
df_betweenness.rename(columns={"id":"country"}, inplace=True)
df_betweenness = df_betweenness.loc[:, ["country", "betweenness"]]
df_betweenness
```

```python
df_CONP.rename(columns={"Countries":"country", "Contencao_Populacional":"pop_cont"}, inplace=True)
df_CONP
```

```python
df_coreness = df_coreness.reset_index()
df_coreness.rename(columns={"id":"country"}, inplace=True)
df_coreness = df_coreness.loc[:, ["country", "coreness"]]
df_coreness
```

```python
df_fluxo_entrada = df_fluxo_entrada.reset_index()
df_fluxo_entrada.rename(columns={"destiny":"country", "value":"in_degree"}, inplace=True)
df_fluxo_entrada
```

```python
df_fluxo_saida = df_fluxo_saida.reset_index()
df_fluxo_saida.rename(columns={"origin":"country", "value":"out_degree"}, inplace=True)
df_fluxo_saida
```

```python
df_metrics = df_betweenness.merge(df_CONP, on="country", how="left").merge(df_coreness, on="country", how="left").merge(df_fluxo_entrada, on="country", how="left").merge(df_fluxo_saida, on="country", how="left")
df_metrics
```

```python
df_final = df3.merge(df_metrics, on="country", how="left")
```
