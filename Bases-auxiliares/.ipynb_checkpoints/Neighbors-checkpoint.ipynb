```python
import pandas as pd
```

```python
df_vizinhos = pd.read_csv("./raw-files/Neighbors.csv")
```

```python
df_vizinhos.head()
```

```python
lista_completa = df_vizinhos.country_border_name.tolist()
```

```python
paises = df_vizinhos.country_border_name.unique().tolist()
paises[:5]
```

```python
dic_contagem_vizinhos = {"country_name": [], "n_neightbours": []}
for pais in paises:
    dic_contagem_vizinhos["country_name"].append(pais)
    dic_contagem_vizinhos["n_neightbours"].append(lista_completa.count(pais))
```

```python
dic_contagem_vizinhos
```

```python
df_contagem_vizinhos = pd.DataFrame(dic_contagem_vizinhos)
```
